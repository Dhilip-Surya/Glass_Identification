```python
import pandas as pd
```

```python
bank = pd.read_csv('clean_data.csv',index_col=0)
```

```python
bank
```

```python
import matplotlib.pyplot as plt
```

# Histogram for Age Attribute

```python
bank['age'].plot(kind='hist',bins=30,figsize=(8,8)) 
plt.title('AGE')
plt.xlabel('AGE')
plt.ylabel('FREQUENCY')
plt.show()
```

# Bar Graph for Jobs Attribute

```python
bank['job'].value_counts().plot(kind='bar',align='center',figsize=(8,8)) 
plt.xlabel('JOBS')
plt.ylabel('FREQUENCY')
plt.title('TYPE OF JOBS')
plt.show()
```

# Pie Chart for Marital Status

```python
bank['marital'].value_counts().plot(kind='pie',autopct='%.2f',startangle=90,figsize=(8,8),legend='4') 
plt.title('MARITAL STATUS')
plt.show()
```

# Bar Graph for Education Attribute

```python
bank['education'].value_counts().plot(kind='bar',align='edge',figsize=(8,8)) 
plt.title('LEVEL OF EDUCATION')
plt.xlabel('EDUCATION TYPE')
plt.ylabel('FREQUENCY')
plt.show()
```

# Bar Graph for Default Attribute

```python
bank['default'].value_counts().plot(kind='bar',figsize=(8,8)) 
plt.xlabel('credit')
plt.ylabel('Frequency')
plt.title('HAS CREDIT IN DETAIL')
plt.show()
```
